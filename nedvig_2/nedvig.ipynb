- низкая стоимость - до 3.4 миллиона; 
- средняя стоимость от 3.4 до 6.8 миллиона; 
- высокая стоимость от 6.8 миллионов руб до 15 миллионов;
- "сверх дорогое жилье" от 15 миллионов до 50 миллионов; 
- от 50 миллионов - лахури.

```python
def last_price_id(last_price):
    if last_price<=3400000:
        return 'низкая стоимость'
    elif last_price<=6800000:
        return "среднеяя стоимость"
    elif last_price<=15000000:
        return "высокая стоимость"
    elif last_price<=50000000:
        return "высокая стоимость"
    else:
        return 'лахури'
```

```python
df['last_price_id']=df['last_price'].apply(last_price_id)
```
